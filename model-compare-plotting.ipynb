```python
# comparison boxplot r squared

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd


data = {
    "Linear Ridge Regression": [0.907, 0.919, 0.918, 0.890, 0.924, 0.885, 0.846, 0.934, 0.920, 0.943, 0.783, 0.900, 0.813, 0.770],
    "Linear Lasso Regression": [0.822, 0.794, 0.757, 0.764, 0.808, 0.735, 0.653, 0.852, 0.792, 0.825, 0.668, 0.751, 0.753, 0.748],
    "Random Forest": [0.924, 0.930, 0.933, 0.908, 0.944, 0.915, 0.895, 0.941, 0.929, 0.959, 0.799, 0.923, 0.850, 0.773],
    "CNN": [0.927, 0.929, 0.903, 0.906, 0.939, 0.908, 0.877, 0.933, 0.921, 0.956, 0.781, 0.907, 0.846, 0.786]
}


stats_lin_ridge = []
stats_lin_lasso = []
stats_random_forest = []
stats_cnn = []

stats_lin_ridge.append({
    'mean': np.mean(data["Linear Ridge Regression"]),
    'median': np.median(data["Linear Ridge Regression"]),
    'std': np.std(data["Linear Ridge Regression"]),
    'max': np.max(data["Linear Ridge Regression"]),
    'min': np.min(data["Linear Ridge Regression"])
})
print(stats_lin_ridge)

stats_lin_lasso.append({
    'mean': np.mean(data["Linear Lasso Regression"]),
    'median': np.median(data["Linear Lasso Regression"]),
    'std': np.std(data["Linear Lasso Regression"]),
    'max': np.max(data["Linear Lasso Regression"]),
    'min': np.min(data["Linear Lasso Regression"])
})
print(stats_lin_lasso)

stats_random_forest.append({
    'mean': np.mean(data["Random Forest"]),
    'median': np.median(data["Random Forest"]),
    'std': np.std(data["Random Forest"]),
    'max': np.max(data["Random Forest"]),
    'min': np.min(data["Random Forest"])
})
print(stats_random_forest)

stats_cnn.append({
    'mean': np.mean(data["CNN"]),
    'median': np.median(data["CNN"]),
    'std': np.std(data["CNN"]),
    'max': np.max(data["CNN"]),
    'min': np.min(data["CNN"])
})
print(stats_cnn)


plt.figure(figsize=(16, 16))
sns.boxplot(data=data)

plt.title('Boxplot Models r squared comparison plot (higher is better)', fontsize=16)
plt.xlabel('Model', fontsize=12)
plt.ylabel('R squared values', fontsize=12)

plt.savefig('model_comparison_plot.png', dpi=300, bbox_inches='tight')

plt.show()
```
